```
import requests
from requests_html import HTMLSession
from bs4 import BeautifulSoup
 
url = "https://www.nytimes.com/"
 
try:
    session = HTMLSession()
    response = session.get(url)
     
except requests.exceptions.RequestException as e:
    print(e)
```

```
title =  response.html.find('title')
print(title)
# [<Element 'title' >]
```

```
print(title[0].text)
```

```
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```
text = soup.find_all(text=True)
```

```
set([t.parent.name for t in text])
```

```
blacklist = [
    '[document]',
    'a',
    'body',
    'button',
    'div',
    'h2',
    'h3',
    'head',
    'html',
    'li',
    'noscript',
    'p',
    'picture',
    'script',
    'span',
    'style'
    # there may be more elements you don't want, such as "style", etc.
]
output = ''
for t in text:
    if t.parent.name not in blacklist:
        output += '{} '.format(t)

print(output)
```

```
title = soup.find('title')

print(title) # Prints the tag
print(title.string) # Prints the tag string content
```

```
for title in soup.find_all('title'): 
    print(title.get_text())
```

```
soup.find_all("h3")
```

```
for header in soup.find_all('h3'): 
    print(header.get_text())
```

```
 headlines = soup.find_all("span", {'class' : 'balancedHeadline'})
```

```
for header in headlines:
    print(header)
```

```
lines = [span.get_text() for span in headlines]
lines
```

```
titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```
